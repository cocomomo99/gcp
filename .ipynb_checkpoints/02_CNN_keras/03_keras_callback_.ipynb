```python
import tensorflow as tf
import numpy as np
import seaborn as sns
```

```python
data = tf.keras.datasets.fashion_mnist.load_data()
data
```

```python
(x_train, y_train), (x_test, y_test) = data
```

```python
x_train.shape, y_train.shape
```

```python
# 이미지 정규화

x_train = x_train / 255.0; x_test = x_test / 255.0

x_train
```

```python
x_train = tf.reshape(x_train, (-1, 28, 28, 1))
x_test = tf.reshape(x_test, (-1, 28, 28, 1))
```

```python
x_train.shape
```

```python
model_VGG = tf.keras.Sequential([
    tf.keras.layers.Conv2D(filters= 32, kernel_size=(3,3), strides= (1,1), padding= 'same', input_shape= (28,28,1), activation='relu'),
    tf.keras.layers.Conv2D(filters= 64, kernel_size=(3,3), strides= (1,1), padding= 'same', activation='relu'),
    tf.keras.layers.MaxPool2D(pool_size= (2,2)),

    tf.keras.layers.Conv2D(filters= 128, kernel_size=(3,3), strides= (1,1), padding= 'same', activation='relu'),
    tf.keras.layers.Conv2D(filters= 256, kernel_size=(3,3), strides= (1,1), padding= 'same', activation='relu'),
    tf.keras.layers.MaxPool2D(pool_size= (2,2)),

    tf.keras.layers.Conv2D(filters= 512, kernel_size=(3,3), strides= (1,1), padding= 'same', activation='relu'),
    tf.keras.layers.Conv2D(filters= 1024, kernel_size=(3,3), strides= (1,1), padding= 'valid', activation='relu'),
    tf.keras.layers.MaxPool2D(pool_size= (2,2)),

    tf.keras.layers.Flatten(),
    tf.keras.layers.Dense(units= 512, activation= 'relu'),
    tf.keras.layers.Dropout(rate = 0.5),
    tf.keras.layers.Dense(units= 256, activation= 'relu'),
    tf.keras.layers.Dropout(rate = 0.5),
    tf.keras.layers.Dense(units= 10, activation= 'softmax')
])


model_VGG.summary()
```

```python
class MyCallback(tf.keras.callbacks.Callback):
    def on_epoch_end(self, epoch, logs = {}):
        if logs.get('accuracy') >= 0.95:
            print('\nReached 90% accuracy so canccelling the trainning')
            self.model.stop_training = True

callbacks = MyCallback()
```

```python
model_VGG.compile(optimizer = 'adam',
                  loss ='sparse_categorical_crossentropy',
                  metrics=['accuracy'])
hist = model_VGG.fit(x_train, y_train, epochs= 25, batch_size= 64, validation_split= 0.2, callbacks = [callbacks])
```

```python
model_VGG.evaluate(x_test, y_test)
```

```python
sns.lineplot(x = np.arange(0, len(hist.history['loss'])), y = hist.history['accuracy'])
```

```python
sns.lineplot(x = np.arange(0, len(hist.history['loss'])), y = hist.history['loss'])
```

```python
import pandas as pd
import tensorflow as tf
import numpy as np
import seaborn as sns

train_df = pd.read_csv(
    'C:\\Users\\HOME\\PycharmProjects\\ai\\.ipynb_checkpoints\\01_tensorflow_basic\\data_set\\caesarian.csv')
print(train_df)
```

```python
x_train = train_df.iloc[ :, :-1 ].to_numpy()
x_train = (x_train - np.min(x_train, axis = 0)) / (np.max(x_train, axis = 0) - np.min(x_train, axis = 0))
y_train = train_df.iloc[ :, -1 ].to_numpy()
x_train.shape, y_train.shape
```

```python
model1 = tf.keras.Sequential(layers=[
    tf.keras.layers.Dense(units=1024, activation='relu', use_bias=True, input_shape=(5,)),
    tf.keras.layers.Dense(units=512, activation='relu', use_bias=True),
    tf.keras.layers.Dense(units=256, activation='relu', use_bias=True),
    tf.keras.layers.Dense(units=128, activation='relu', use_bias=True),
    tf.keras.layers.Dense(units=64, activation='relu', use_bias=True),
    tf.keras.layers.Dense(units=32, activation='relu', use_bias=True),
    tf.keras.layers.Dense(units=1, activation='sigmoid', use_bias=True)
])

model1.summary()
```

```python
class MyCallback(tf.keras.callbacks.Callback):
    def on_epoch_end(self, epoch, logs = {}):
        if logs.get('loss') <= 0.53:
            print('\n>>>> Reached 0.53 >= loss so canccelling the trainning')
            self.model.stop_training = True
MyCallback = MyCallback()
```

```python
model1.compile(loss='binary_crossentropy', optimizer='adam', metrics= ['accuracy'])
hist1 = model1.fit(x=x_train, y=y_train, epochs=300, validation_split=0.2, batch_size = 3,
                   # callbacks = [MyCallback]
                   )
```

```python
model1.evaluate(x_train, y_train)
```

```python
sns.lineplot(x=range(0, len(hist1.history['loss'])), y=hist1.history['accuracy'])
```

```python
train_df = pd.read_csv(
    'C:\\Users\\HOME\\PycharmProjects\\ai\\.ipynb_checkpoints\\01_tensorflow_basic\\data_set\\iris.csv')
print(train_df)
test_df = pd.read_csv(
    'C:\\Users\\HOME\\PycharmProjects\\ai\\.ipynb_checkpoints\\01_tensorflow_basic\\data_set\\iris.csv')
print(test_df)
```

```python
x_train = train_df.iloc[ :, :-1 ].to_numpy()
y_train = pd.get_dummies(train_df.iloc[ :, -1 ]).to_numpy()
x_train.shape, y_train.shape
```

```python
model2 = tf.keras.Sequential(layers=[
    tf.keras.layers.Dense(units=40, activation='relu', use_bias=True, input_shape=(5,)),
    tf.keras.layers.Dense(units=20, activation='relu', use_bias=True),
    tf.keras.layers.Dense(units=3, activation='sigmoid', use_bias=True)
])

model2.summary()
```

```python
class MyCallback2(tf.keras.callbacks.Callback):
    def on_epoch_end(self, epoch, logs = {}):
        if logs.get('accuracy') >= 0.9:
            print('\nReached 0.9 accuracy so canccelling the trainning')
            self.model.stop_training = True
MyCallback = MyCallback2()
```

```python
model2.compile(loss='categorical_crossentropy', optimizer='adam', metrics=['accuracy'])
hist2 = model2.fit(x=x_train, y=y_train, epochs=700, batch_size = 10, callbacks = [MyCallback])
```

```python
sns.lineplot(x=range(0, len(hist2.history[ 'loss' ])), y=hist2.history[ 'loss' ])
```

```python
sns.lineplot(x=range(0, len(hist2.history[ 'accuracy' ])), y=hist2.history[ 'accuracy' ])
```
