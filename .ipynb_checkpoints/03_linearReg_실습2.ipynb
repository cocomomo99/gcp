```python
import tensorflow as tf
import pandas as pd
import numpy as np
from plotnine import *

tf.random.set_seed(5)
```

```python
boston_train = pd.read_csv('C:/Users/HOME/PycharmProjects/ai/.ipynb_checkpoints/data_set/boston_train.csv')
boston_test = pd.read_csv('C:/Users/HOME/PycharmProjects/ai/.ipynb_checkpoints/data_set/boston_test.csv')
boston_train
```

```python
x_train = boston_train.iloc[:,:-1].to_numpy()
x_train
```

```python
y_train = boston_train.iloc[:,-1].to_numpy()
y_train
```

```python
w = tf.Variable(tf.random.normal([9,1], dtype= tf.float64), name= 'weight')
b = tf.Variable(tf.random.normal([1], dtype= tf.float64), name= 'bias')
w, b
```

```python
def cost_funct():
    y_pred = tf.matmul(x_train, w) + b
    loss = tf.reduce_mean(tf.square(y_train - y_pred))
    return loss
```

```python
cost_funct()
```

# 학습시작

```python
optimizer = tf.optimizers.Adam(learning_rate=0.01)

print('***** Start Learning!!')

for step in range(10001):
    optimizer.minimize(loss=cost_funct, var_list=[w, b])
    if step % 100 == 0:
        print('%04d' % step, 'cost:[', cost_funct().numpy(), ']',
              ' W:', w.numpy(), ' b:', b.numpy())
        print('1')
print('***** Learning Finished!!')
```

# 정확도 측정

```python
def pred(x, w, b):
    pred = tf.matmul(x, w) + b
    return pred

# 정확도 측정 : RMSE(Root Mean Squared Error)
def get_rmse(y_test,preds):
    squared_error = 0
    for k,_ in enumerate(y_test):
        squared_error += (preds[k] - y_test[k])**2
    mse = squared_error/len(y_test)
    rmse = np.sqrt(mse)
    return rmse[0]
```

```python
# 학습한 데이터를 그대로 검증 데이터로 사용한 경우
y_test = boston_test.iloc[:, -1]
x_test = boston_test.iloc[:, :-1]

preds = pred(x_test, w, b)
print('RMSE:',get_rmse(y_test,preds))  # RMSE:
```
